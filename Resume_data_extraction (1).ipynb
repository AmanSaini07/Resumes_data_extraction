```python
import os
import re
import csv
from PyPDF2 import PdfReader

# Set the directory path where the PDF files are located
dir_path = 'C:/Users/MR.AMAN SAINI/OneDrive/Desktop/resume'

# Set up regular expressions for data extraction
registration_number_pattern = re.compile(r'project registration number\s*:\s*(\w+)', re.IGNORECASE)
project_pattern = re.compile(r'project\s*:\s*(.+)', re.IGNORECASE)
plot_pattern = re.compile(r'plot\s*bearing\s*/\s*cts\s*/\s*survey\s*/\s*final\s*plot\s*no\s*.\s*:\s*(.+)', re.IGNORECASE)
address_pattern = re.compile(r'at\s*(.+)', re.IGNORECASE)
validity_pattern = re.compile(r'commencing\s*from\s*(\d{2}/\d{2}/\d{4})\s*and\s*ending\s*with\s*(\d{2}/\d{2}/\d{4})', re.IGNORECASE)

# Set up the CSV writer
csv_file = open('C:/Users/MR.AMAN SAINI/OneDrive/Desktop/resume/resume_data.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['File', 'Registration Number', 'Project', 'Plot', 'Address', 'Validity'])

# Loop through the PDF files in the directory
for filename in os.listdir(dir_path):
    if filename.endswith('.pdf'):
        # Build the full file path
        file_path = os.path.join(dir_path, filename)

        # Open the PDF file with PyPDF2
        with open(file_path, 'rb') as pdf_file:
            # Create a PDF reader object
            pdf_reader = PdfReader(pdf_file)

            # Extract text from each page of the PDF
            text = ''
            for page in pdf_reader.pages:
                text += page.extract_text()

            # Apply regular expressions to extract the data
            registration_number = re.search(registration_number_pattern, text)
            project = re.search(project_pattern, text)
            plot = re.search(plot_pattern, text)
            address = re.search(address_pattern, text)
            validity = re.search(validity_pattern, text)

            # Write the extracted data to the CSV file
            row = [filename]
            if registration_number:
                row.append(registration_number.group(1))
            else:
                row.append('')
            if project:
                row.append(project.group(1))
            else:
                row.append('')
            if plot:
                row.append(plot.group(1))
            else:
                row.append('')
            if address:
                row.append(address.group(1))
            else:
                row.append('')
            if validity:
                row.append(validity.group(1) + ' to ' + validity.group(2))
            else:
                row.append('')
            csv_writer.writerow(row)

# Close the CSV file
csv_file.close()
```
