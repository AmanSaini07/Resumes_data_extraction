# For converting docx or txt to pdf's


```python
import os
from docx2pdf import convert
from fpdf import FPDF
import time
# Set the directory containing the files you want to convert
source_directory = 'A:/doc1 files'

# Set the directory where you want to save the converted files
output_directory = 'A:/resumes pdf'

# Loop through all the files in the directory
for filename in os.listdir(source_directory):
    if filename.endswith('.docx') or filename.endswith('.txt'):
        # Convert the file to pdf using docx2pdf or fpdf
        if filename.endswith('.docx'):
            convert(os.path.join(source_directory, filename), output_directory)
            filename = filename.replace('.docx', '.pdf')
        elif filename.endswith('.txt'):
            pdf = FPDF()
            pdf.add_page()
            with open(os.path.join(source_directory, filename), 'r') as f:
                text = f.read()
            pdf.cell(0, 10, text)
            filename = filename.replace('.txt', '.pdf')
            pdf.output(os.path.join(output_directory, filename))

print('Conversion complete!')
```

# Extraction of Contact Numbers


```python
import os
import re
import pandas as pd
from PyPDF2 import PdfReader

# Define the regular expression pattern to match phone numbers
phone_pattern = re.compile(r'\d{12}|\d{10}')

# Define the file paths
pdf_folder_path = 'A:/resumes pdf'
xlsx_file_path = 'A:/Resumes_data/contacts_digits.xlsx'

# Create empty lists to store the extracted phone numbers and file names
phone_numbers = []
file_names = []

# Loop through each file in the PDF folder
for filename in os.listdir(pdf_folder_path):
    if filename.endswith('.pdf'):
        file_path = os.path.join(pdf_folder_path, filename)
        # Open the PDF file in read-binary mode
        with open(file_path, 'rb') as pdf_file:
            # Create a PDF reader object
            pdf_reader = PdfReader(pdf_file)
            # Loop through each page in the PDF file
            for page_num in range(len(pdf_reader.pages)):
                # Extract the text from the page
                page_text = pdf_reader.pages[page_num]
                text = page_text.extract_text()
                # Find all phone numbers in the text using the regular expression pattern
                numbers_on_page = re.findall(phone_pattern, text)
                # Add the phone numbers and file name to the lists
                if len(numbers_on_page) > 0:
                    phone_numbers.extend(numbers_on_page)
                    file_names.extend([filename] * len(numbers_on_page))
                else:
                    phone_numbers.append(None)
                    file_names.append(filename)

# Create a pandas DataFrame with the extracted phone numbers and file names
df = pd.DataFrame({'File_name': file_names,'Phone Number': phone_numbers})

# Save the DataFrame to an Excel file
df.to_excel(xlsx_file_path, index=False)
```

# Extraction of Birth dates


```python
import os
import PyPDF2
import re
import csv
import time
import warnings
warnings.filterwarnings('ignore', '.*do not.*', )

# Set the directory path where the PDF files are located
dir_path = 'A:/resumes pdf'
dob_pattern = re.compile(r'Date of (?:B|b)irth\s*[:|\.]?\s*(\d{1,2}(?:st|nd|rd|th)?(?:\s+(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)|(?:\d{2})?[/-](?:0?[1-9]|[12][0-9]|3[01])[/-](?:19|20)\d{2}|\b(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\b\.\s*(?:19|20)?\d{2}))(?:\D+(\d{4}))?')


# Set up the CSV writer
csv_file = open('A:/Resumes_data/dob_data.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['File', 'Date of Birth', 'Year'])

# Loop through the PDF files in the directory
for filename in os.listdir(dir_path):
    if filename.endswith('.pdf'):
        # Build the full file path
        file_path = os.path.join(dir_path, filename)
        
        try:
            # Open the PDF file in read-binary mode
            with open(file_path, 'rb') as pdf_file:
                # Create a PDF reader object
                pdf_reader = PyPDF2.PdfReader(pdf_file)
                
                # Loop through each page in the PDF file
                for page_num in range(len(pdf_reader.pages)):
                    # Extract the text from the page
                    page = pdf_reader.pages[page_num]
                    text = page.extract_text()
                    dob = re.search(dob_pattern, text)
                    
                    # Write the extracted data to the CSV file
                    if dob:
                        row = [filename, dob.group(1), dob.group(2) if dob.group(2) else '']
                        print(row)
                        csv_writer.writerow(row)
                    time.sleep(1)
                        
        except Exception as e:
            # Handle the error and continue with the next file
            print(f"Error processing {filename}: {str(e)}")
            continue

# Close the CSV file
csv_file.close()
```

# Extraction of Gender

```python
import os
import PyPDF2
import re
import csv

# Set the directory path where the PDF files are located
dir_path = 'A:/resumes pdf'
gender_pattern = re.compile(r'Gender\s*:\s*(\w+)', re.IGNORECASE)

# Set up the CSV writer
csv_file = open('A:/Resumes_data/gender_data.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['File', 'Gender'])

# Loop through the PDF files in the directory
for filename in os.listdir(dir_path):
    if filename.endswith('.pdf'):
        # Build the full file path
        file_path = os.path.join(dir_path, filename)
        
        # Open the PDF file in read-binary mode
        with open(file_path, 'rb') as pdf_file:
            # Create a PDF reader object
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            
            # Loop through each page in the PDF file
            for page_num in range(len(pdf_reader.pages)):
                # Extract the text from the page
                page = pdf_reader.pages[page_num]
                text = page.extract_text()
                gender_match = re.search(gender_pattern, text)
                
                # Write the extracted data to the CSV file
                if gender_match:
                    row = [filename]
                    row.append(gender_match.group(1))
                    print(gender_match.group(1))
                    csv_writer.writerow(row)

# Close the CSV file
csv_file.close()
```

# Extarction of Experience

```python
import os
import re
import csv
import PyPDF2

# Set the directory path where the PDF files are located
dir_path = 'A:/resumes pdf'

# Set up the CSV writer
csv_file = open('A:/Resumes_data/experience.csv', 'w', newline='', encoding='utf-8')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['File', 'Experience'])

# Define the regular expression patterns for experience and stop keywords
exp_pattern = re.compile(r'(experience|employment history|professional experience|work experience|Industry Exposure|employment record)', re.IGNORECASE)
stop_pattern = re.compile(r'(IT Skills|Skills|Education|Qualification|Personal Details|Certifications|Degree|Certificate|Personal Profile|Hobbies|Strength|Personal Vitae)', re.IGNORECASE)

# Loop through the PDF files in the directory
for filename in os.listdir(dir_path):
    if filename.endswith('.pdf'):
        # Build the full file path
        file_path = os.path.join(dir_path, filename)
        
        # Open the PDF file in read-binary mode
        with open(file_path, 'rb') as pdf_file:
            # Create a PDF reader object
            pdf_reader = PyPDF2.PdfReader(pdf_file)
            
            # Loop through each page in the PDF file
            for page_num in range(len(pdf_reader.pages)):
                # Extract the text from the page
                page = pdf_reader.pages[page_num]
                text = page.extract_text()
                
                # Search for experience using the regular expression pattern
                exp_matches = exp_pattern.findall(text)
                
                # Search for stop keywords
                stop_matches = stop_pattern.findall(text)
                
                # Write the extracted data to the CSV file
                if exp_matches:
                    start = text.find(exp_matches[0])
                    end = text.find('\n\n', start)
                    
                    # Check if there are any stop keywords before the end of the relevant part
                    for stop_match in stop_matches:
                        stop_index = text.find(stop_match, start, end)
                        if stop_index != -1:
                            end = stop_index
                            break
                    
                    relevant_part = text[start:end].strip()
                    row = [filename, relevant_part]
                else:
                    row = [filename, '']
                csv_writer.writerow(row)
                
# Close the CSV file
csv_file.close()
```
